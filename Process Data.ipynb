```python
def combineData():
    BASE_DATA_DIR = 'Data'
    df_data = pd.DataFrame()
    for data_file in os.listdir(BASE_DATA_DIR):
        if data_file.endswith(".fits"):
            hdulist = fits.open(os.path.join(BASE_DATA_DIR, data_file))
            subclass = hdulist[2].data[0]['SUBCLASS']
            data = hdulist[1].data
            flux = data['flux']
            flux = pd.DataFrame(np.array(flux).byteswap().newbyteorder())
            cols = pd.DataFrame(np.array(data['loglam']).byteswap().newbyteorder())
            flux = flux.transpose()
            cols = cols.transpose()
            cols = cols.values.tolist()
            flux.columns = cols
            flux["SUBCLASS_LETTER"] = subclass[0]
            flux["SUBCLASS_NUMBER"] = subclass[1]
            df_data = df_data.append(flux, ignore_index = True)
            hdulist.close()
    return df_data
```

```python
data = combineData()
```
